```python
import json

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
dqn_results = json.load(open("results/dqn_results.json"))
ppo_results = json.load(open("results/ppo_results.json"))

dqn_rewards_per_episode = dqn_results["episode_rewards"]
ppo_rewards_per_episode = ppo_results["episode_rewards"]

dqn_system_mean_waiting_time_per_episode = dqn_results["system_mean_waiting_time_per_episode"]
ppo_system_mean_waiting_time_per_episode = ppo_results["system_mean_waiting_time_per_episode"]

dqn_t_stopped_vehicles_per_episode = dqn_results["t_stopped_vehicles_per_episode"]
ppo_t_stopped_vehicles_per_episode = ppo_results["t_stopped_vehicles_per_episode"]

sns.set(style="darkgrid")
```

```python
# episode_rewards comparison
plt.figure(figsize=(10,5))
smoothing_window = 20
plt.plot([np.mean(dqn_rewards_per_episode[i:i+smoothing_window]) for i in range(len(dqn_rewards_per_episode)-smoothing_window)], linewidth=1, color="red", label="DQN")
plt.plot(dqn_rewards_per_episode, linewidth=1, color="red", alpha=0.25)
plt.plot([np.mean(ppo_rewards_per_episode[i:i+smoothing_window]) for i in range(len(ppo_rewards_per_episode)-smoothing_window)], linewidth=1, color="blue", label="PPO")
plt.plot(ppo_rewards_per_episode, linewidth=1, color="blue", alpha=0.25)
plt.xlabel("Episode")
plt.ylabel("Episode reward")
plt.legend()
plt.savefig("output/episode_rewards.png")
```

```python
# system_mean_waiting_time_per_episode comparison
plt.figure(figsize=(10, 5))
smoothing_window = 20
plt.plot([np.mean(dqn_system_mean_waiting_time_per_episode[i:i+smoothing_window]) for i in range(len(dqn_system_mean_waiting_time_per_episode)-smoothing_window)], linewidth=1, color="red", label="DQN")
plt.plot(dqn_system_mean_waiting_time_per_episode, linewidth=1, color="red", alpha=0.25)
plt.plot([np.mean(ppo_system_mean_waiting_time_per_episode[i:i+smoothing_window]) for i in range(len(ppo_system_mean_waiting_time_per_episode)-smoothing_window)], linewidth=1, color="blue", label="PPO")
plt.plot(ppo_system_mean_waiting_time_per_episode, linewidth=1, color="blue", alpha=0.25)
plt.xlabel("Episode")
plt.ylabel("System mean waiting time per episode")
plt.legend()
plt.savefig("output/system_mean_waiting_time_per_episode.png")
```

```python
# t_stopped_vehicles_per_episode comparison
plt.figure(figsize=(10, 5))
smoothing_window = 20
plt.plot([np.mean(dqn_t_stopped_vehicles_per_episode[i:i+smoothing_window]) for i in range(len(dqn_t_stopped_vehicles_per_episode)-smoothing_window)], linewidth=1, color="red", label="DQN")
plt.plot(dqn_t_stopped_vehicles_per_episode, linewidth=1, color="red", alpha=0.25)
plt.plot([np.mean(ppo_t_stopped_vehicles_per_episode[i:i+smoothing_window]) for i in range(len(ppo_t_stopped_vehicles_per_episode)-smoothing_window)], linewidth=1, color="blue", label="PPO")
plt.plot(ppo_t_stopped_vehicles_per_episode, linewidth=1, color="blue", alpha=0.25)
plt.xlabel("Episode")
plt.ylabel("T stopped vehicles per episode")
plt.legend()
plt.savefig("output/t_stopped_vehicles_per_episode.png")
```
